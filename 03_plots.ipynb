```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python
air_quality = pd.read_csv('data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality.head()
```

### `plot()`

```python
print(air_quality.shape)
air_quality.plot()
```

```python
air_quality["station_paris"].plot()
```

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
```

### plot function (except default line)

```python
[
    method_name for method_name in dir(air_quality.plot)
    if not method_name.startswith('_')
]
```

```python
# box plot
air_quality.plot.box()
```

```python
axs = air_quality.plot.area(figsize=(12, 4), subplots=True)
```

```python
fig, axs = plot.subplots(figsize=(12, 4))        # Create an empty matplotlib Figure and Axes
air_quality.plot.area(ax=axs)                   # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("NO$_2$ concentration")          # Do any matplotlib customization you like
fig.savefig("result/no2_concentrations.png")    # Save the Figure/Axes using the existing matplotlib method.
```
