```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')

titanic.head()
```

```python
### Shape
titanic["Age"].shape
```

```python
age_sex = titanic[["Age", "Sex"]]
print(age_sex.shape)
age_sex.head()
```

### filter specific rows from a DataFrame

```python
above_35 = titanic[titanic["Age"] > 35]
above_35.head()
```

### `isin`

```python
class_23 = titanic[titanic['Pclass'].isin([1,2])]
class_23.head()
```

### `notna`

```python
age_not_na = titanic[titanic['Age'].notna()]
age_not_na.head()
```

### `iloc` `loc`

```python
adult_names = titanic.loc[titanic['Age'] > 35, ('Name', 'Age')]
print(adult_names.head())

titanic_iloc = titanic.iloc[9:25, 2:5]
print(titanic_iloc.head())
```

```python
titanic.iloc[0:3, 3] = "anonymous"
titanic.head()
```
